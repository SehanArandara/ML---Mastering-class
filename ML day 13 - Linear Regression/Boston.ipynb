## File imports in here 

```python
import pandas as pd    # we have to manipulate the data 
import numpy as np # for numerical operations 
from sklearn.model_selection import train_test_split   # this will split the data set into two parts train and test 
from sklearn.linear_model import LinearRegression  # create a linear regression model
from sklearn.metrics import mean_squared_error  # this function evaluate the actual predicted value and the true test value 
import matplotlib.pyplot as plt
```

## Import the data set 

```python
data = pd.read_csv("C:\\Users\\sdara\\Desktop\\HousingData.csv")
```

```python
data.head()   # we can check the data set  first 5 rows 
```

## step 03 :  Choose the  x and Y and convert these tabulars formates to 2D arrays 

```python
x = data.iloc[:,:12].values    # input 
y = data.iloc[:,12].values   # out put
# value convert data into arrays 

```

## step 04 :  Split the data set into train and test data 

- train_test_split(x,y,test_size=0.2)   -> this method have three parameters x , y and the size of test test data set 0.2 means 20%
- The above method will return the 4 arrays . They are x train , x test , y train , y test
- The function will divide the data set in random way so we have to stop that we will use the random_state = 0 for that

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=3)
x_train
```

## Train the model by giving the x and y to the model 

```python
model = LinearRegression() 
```

```python
model.fit(x_train,y_train)
```

## The equation for multiple linear regression is: 

Where:

- y is the dependent variable (output/predicted value).
- x_1, x_2, ..., x_n are the independent variables (inputs).
- β_0 is the intercept (constant term).
- β_1, β_2, ..., β_n are the coefficients (weights) for each independent variable.
- ε is the error term (the residual difference between the predicted and actual values).

```python
model.intercept_    
```

```python
model.coef_ 
```

##  Measures how well the model's predictions fit the actual data. 

Purpose:
- R^2 score indicates how well the independent variables explain the variability in the dependent variable.
- A value of 1 means perfect prediction (the model explains all the variability in the data).
- A value of 0 means the model explains no variability (equivalent to just predicting the mean).
- Negative values indicate that the model performs worse than a horizontal line (predicting the mean of the target).

```python
model.score(x_train,y_train)*100    # convert it to the Perecentage 
```

## Get the predicted values by using the x_test data set 

## Calculate the Mean Square Error

- MSE measures how far the predicted values are from the actual values. It averages the square of the differences between predicted and actual values, giving a sense of how much error exists between the predictions and the real data.
- Lower MSE indicates better model performance.

## Get the Root Mean Squared Error - RMSE 
- RMSE gives an error metric in the same units as the target variable (since it's the square root of the MSE).
- It provides an intuitive understanding of how much error the model is making, in terms of the actual output variable's scale.


```python
y_pred = model.predict(x_test)
```

```python
MSE = mean_squared_error(y_pred,y_test)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```
